```python
import pandas as pd

df = pd.read_csv("btc_cleaned.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["date"], df["price"])
plt.title("Bitcoin Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["date"], df["volume"])
plt.title("Bitcoin Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["date"], df["market_cap"])
plt.title("Bitcoin Market Capitalization Over Time")
plt.xlabel("Date")
plt.ylabel("Market Cap")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["price"], bins=30)
plt.title("Distribution of Bitcoin Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
df[["price", "market_cap", "volume"]].corr()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df[["price", "market_cap", "volume"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
